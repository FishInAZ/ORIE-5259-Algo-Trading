Idea 1: Order Book Imbalance + Microprice Timing

Short-term price movements are heavily driven by supply/demand imbalance in the order book. If bid-side pressure is stronger, price is more likely to move up soon (and vice versa).

Therefore:
Buy algorithm → wait for downward pressure (buy cheaper)
Sell algorithm → wait for upward pressure (sell higher)

Within each minute:
Continuously monitor signals and define a score:
score = w1 * OBI + w2 * (microprice − mid) + w3 * momentum

Buy strategy:
Execute when score is most negative (downward pressure), or at last second if no good signal
Sell strategy:
Execute when score is most positive
