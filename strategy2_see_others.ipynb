My idea uses cross-stock lead-lag effects to improve execution timing.
The intuition is that related large-cap stocks often do not move simultaneously at the second-by-second level.
If peer stocks have already moved but the target stock has not fully adjusted yet, the target may follow shortly after.
So instead of executing mechanically like TWAP, I use peer stock movements as a short-term predictive signal to decide whether to trade immediately or wait within the minute.
